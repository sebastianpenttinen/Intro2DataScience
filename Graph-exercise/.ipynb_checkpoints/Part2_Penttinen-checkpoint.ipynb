### Problem Description


Miniproject 2 is about getting familjar with Network Analysis and using NetworkX in doing so.

The data that the network is built from is small/medium-scale protein-protein interactions.

What we want to do with the data is obviously build the network and then run some basic analysis on it.

Like:

1. Figuring out the amount of nodes edges and average degree.

2. Find out what the density of the network is.

3. Find the minimum spanning tree is also of interest.

4. Plot the degree histogram.

5. Find the largest component of the network.
        



```python
# Imports
import networkx as nx
import matplotlib.pyplot as plt
import collections
from matplotlib.pyplot import figure
```

```python
# load the data
with open('CE-GT.txt') as f:
    content = f.readlines()
content = [x.strip() for x in content]
tupleList = []

for element in content:
    tupleList.append(tuple(filter(None, element.split('	'))))
weights = [float(x[2]) for x in tupleList] 
node1 = [x[0] for x in tupleList]
node2 = [x[1] for x in tupleList]
```

I assumed that the graph was undirected.

```python
# Make the Graph
%matplotlib inline
G = nx.Graph()
for i in range(0,len(node1)):
    G.add_edge(node1[i],node2[i],weight=weights[i])
nx.draw(G)
```

Here is how it would look if it was directed.

Note that the nodes and so on wount chage if it is directed or not, but the edges will be drawn in the plot.

```python
K = nx.DiGraph()
for i in range(0,len(node1)):
    K.add_edge(node1[i],node2[i],weight=weights[i])
nx.draw(K)
```

```python
# calculate number of nodes
print('There are '+ str(G.number_of_nodes()) +' nodes in the network')
```

```python
print('Thera are '+str(G.number_of_edges()) + ' edges in the network')
```

```python
sortedDegrees = degreeSequence = sorted([x for i, x in G.degree()], reverse=True) # sorting is needed later for the histogram 
print('The average degree of the network is: ' + str(sum(sortedDegrees)/len(sortedDegrees)))
```

```python
print('The density of the network is: ' + str(nx.density(G)))
```

```python
minimumSpan =  nx.minimum_spanning_tree(G) # Can't make a minimum spannig tree of a directed graph
nx.draw(minimumSpan)
```

# Degree histograms
This "histogram" is made with a barplot since it shows the smaller changes better and according to the example provided in the NetworkX documentation

```python
figure(num=None, figsize=(10, 15), dpi=100, facecolor='w', edgecolor='k')
```

```python
degreeCount = collections.Counter(sortedDegrees)
degree, count = zip(*degreeCount.items())
fig, ax = plt.subplots()
plt.bar(degree, count, width=1.0)
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.show()
```

# Find the largest connected component of the network

```python
LC = max(nx.connected_component_subgraphs(G), key=len)
nx.draw(LC)
```

```python
print('The diameter if the LC is: '+ str(nx.diameter(LC)))
```

```python
print('The center of the LC is: ' + str(nx.center(LC)))
```

```python
numberOfCliques =  nx.number_of_cliques(G)
nodesWith3 = []
for key,value in numberOfCliques.items():
    if value == 3:
        nodesWith3.append(key)
print('Number of clique communities with 3 nodes: '+str(len(nodesWith3)))
```
